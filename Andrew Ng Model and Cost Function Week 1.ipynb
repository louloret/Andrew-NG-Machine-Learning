## Addition and Scalar Multiplication

```python
Image(filename = path + "addition_matrix.png")
```

```python
# Initialize matrix A and B
A = np.array([[1, 2, 4],
              [5, 3, 2]])

B = np.array([[1, 3, 4],
              [1, 1, 1]])

#Initialize constant s
s = 2
```

```python
# see how element-wise edition works
add_AB = A + B
add_AB
```

```python
#See how element-wise subtraction works
sub_AB = A - B
sub_AB
```

```python
#See how scalar multiplication works
mult_As = A * s
mult_As
```

```python
#Divide A by s
div_As = A / s
div_As
```

```python
# What happens if we have a Matrix + scalar?
add_As = A + s
add_As
#each element is added by 2
```
