### how to distinguish between high bias and high variance problems

#### high bias aka underfit example

```python
##increasing training set size 
Image(filename =  path + "high_bias.png")
```

```python
## cv error will plateu pretty soon when fitting a straight line
## training error will get converge very close to cv error, due to so few
## parameters and such large data that the performance between cv and train
# data are very similar
```

```python
#notice that increasing size of data will not (by itself) improve performance
# when there is high bias in algorithm, this is a useful thing to know
# in order not to waste time collecting more data
```

#### high variance aka overfitting example

```python
## as training data increases, error rate increases because it is harder
# to fit all that data perfectly
Image(filename =  path + "high_variance.png")
```

```python
## cv error will decrease a bit as data increases but remain very high

# a large gap between cv error and training error is a signal of high variance
# error
```

```python
## in this case more data will likely help model performance aka cv error, which
# is what we really care about
```

```python
## this will signal that it is actually useful to get more data
```
