## Mathematical Intuition behind Normal Equation

```python
Image(filename = path + "least_squares.png")
```

```python
# no solution for x1a1 + x2a2 + .... + xnan = vector b
# vector b is not in column space of A
# there is no solution, but we want to get as close to vector b as possible
# find x* as close to b as possible
# min(length(vector b - Ax*))
```

```python
Image(filename = path + "analytically.png")
```

```python
# this why this is the Least Squares Approximation
# the vector in subspace of A that is closest to vector b
# just the projection of vector b onto the subspace of A
# Ax* = proj vector b onto subspace of A
```

#### What happens if we subtract b from both sides of equation?

```python
Image(filename = path + "orthogenal.png")
```

```python
# vector(proj b - vector b) is orthogenal to subspace of A by definition
# vector(proj b - vector b) is orthogenal complment of subspace
#  orthogenal complement of subspace = null space of A transpose
# Ax* - vector b is a member of null space of A transpose
# this means if i multiply A transpose by (Ax* - vector b) = 0
# distribute and move to other side and you are left with normal equation
```

```python
Image(filename = path + "step1.png")
```

#### Why does this algebra work?

```python
# we are trying to find x* that minimizes (vector b - Ax*) aka least square solution
# to find this is to find closest vector in our subspace to vector b
# this is the projection of b onto our subspace
# Ax* = Proj vector b onto A
# we need to find x* that satisfies this equation
# we have seen before the the projection b is hard
# simpler way is to convey vector b in subspace A by multiplying by A transpose
# this can be done by multiplying both sides by A transpose
# so now we have: A transpose times Ax* = A transpose times (vector b)
# so now we need to find x* which satisfies this expression of vector b 
```
